```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
house_price_dataframe = pd.DataFrame(data,columns=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT'])
```

```python
#print first 5 rows of datafrme
house_price_dataframe.head()
```

```python
#add the target (price ) column to the data frame
house_price_dataframe['price'] = target
```

```python
#Understanding the correlation between various featurs in the dataset
correlation = house_price_dataframe.corr()
```

```python
#Constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square = True,fmt ='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

```python
#Splitting the data and Target
X = house_price_dataframe.drop(['price'],axis=1)#axis = 1 for columns drop
Y = house_price_dataframe['price']
```

```python
print(X)
print(Y)
```

```python
#Splitting the data into training data and Test data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Training the Model

XGBoost Regressor

```python
#loading the model
model = XGBRegressor()
```

```python
#training the model with X_train
model.fit(X_train,Y_train)
```

Evaluation

Prediction on training data

```python
#accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train,training_data_prediction	)

#Mean absolute error
score_2 = metrics.mean_absolute_error(Y_train,training_data_prediction)

print("R squared :", score_1)
print("Mean absolute error :", score_2)
```

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual prices")
plt.ylabel("Predicted Prices")
plt.title("Actualvs Predicted")
plt.show()
```

Prediction on test data


```python
#accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test,test_data_prediction	)

#Mean absolute error
score_2 = metrics.mean_absolute_error(Y_test,test_data_prediction)

print("R squared :", score_1)
print("Mean absolute error :", score_2)
```

```python
# graph
plt.scatter(Y_test,test_data_prediction)
plt.xlabel("Actual prices")
plt.ylabel("Predicted values")
plt.title("Actual vs Predicted")
plt.show()
```
